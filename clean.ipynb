## Playing with the pokemon API

```python
import requests as r # often requests is imported under the alias r
result = r.get('https://pokeapi.co/api/v2/pokemon/')
```

```python
class Pokedex:
    def __init__(self):
        self.pokemon = {}
        
    def add_poke(self, pname):
        result = r.get("https://pokeapi.co/api/v2/pokemon/" + pname)
        data = result.json()
        w = data['weight']
        new_poke = Pokes(pname, w)
        for a in data['abilities']:
            new_poke.abilities.append(a['ability']['name'])
        for b in data['types']:
            new_poke.types.append(b['type']['name'])
        self.pokemon[pname] = new_poke
        
    def build_lib(self):
        result = r.get("https://pokeapi.co/api/v2/pokemon/")
        data = result.json()
        poke_names = []
        for n in data['results']:
            poke_names.append(n['name'])
        for p in poke_names:
            self.add_poke(p)
            
    def show_dex(self):
        for n in self.pokemon:
            self.pokemon[n].disp_poke()
            
    
    
#     def search_dex(self, pname):
#         self.pname = pname
#         for n in self.pokemon:
#             if self.pokemon[pname] == pname:
#                print(f'{self.pokemon[pname].name}'')
        
class Pokes:
    def __init__(self, name, weight):
        self.abilities = []
        self.types = []
        self.name = name
        self.weight = weight
        
    def disp_poke(self):
        print(f"\n{self.name.title()}- Weight- {self.weight}.\n Types= {self.types}\n Abilities= {self.abilities}")
        
        
```

```python
the_dex = Pokedex()
the_dex.build_lib()
```

```python
the_dex.show_dex()
```
